### Normal request method dont work on websites like youtube and even on most of the websites  that's why we have to use a webdriver    

A WebDriver is like a remote control for your web browser.

Imagine your Chrome, Firefox, or Edge browser is a car 

You can drive it manually (click, type, scroll).

Or you can use a remote control to tell it exactly what to do.

A WebDriver is that remote control — it lets your Python script open a browser window, click buttons, fill out forms, wait for pages to load, and grab HTML, just like you do by hand.

 

**Requests and BeautifulSoup can’t:**

Click buttons,

Run JavaScript,

Log in through pop-ups,

Wait for pages to finish rendering dynamic content.

A WebDriver can do all that because it actually runs a real browser in the background, so the page loads exactly like it does when you open Chrome yourself.

 How it works — 

1️⃣ You install a WebDriver for your browser, like:

**chromedriver for Chrome**

**geckodriver for Firefox.**

2️⃣ You use a tool like **Selenium** or **Playwright** to tell that WebDriver what to do.

3️⃣ The WebDriver opens a real browser window for you, does all the clicks & typing, and returns the page’s final HTML — including content loaded by JavaScript!

```python

soup = BeautifulSoup(driver.page_source , 'html.parser')
sp = soup.find_all('ytd-rich-item-renderer')
data = []
for sp in soup.find_all('ytd-rich-item-renderer'):
    title         = sp.find('h3').text
    video_link    = 'https://www.youtube.com'+sp.find('a').get('href')
    views         = sp.find_all('span' , class_ = 'inline-metadata-item style-scope ytd-video-meta-block')[0].text
    date_time     = sp.find_all('span' , class_ = 'inline-metadata-item style-scope ytd-video-meta-block')[1].text
    try:
        thumbnail_link = sp.find('img').get('src').split('?')[0]
    except:
        thumbnail_link = np.nan    
    data.append([title , views ,date_time ,video_link,thumbnail_link])
df = pd.DataFrame(data , columns = ['title','views','date_time','video_link','thumbnail_link'] )
df.to_csv('data(3).csv' , index = False)
```

```python
# Search for the ytInitialPlayerResponse JSON
pattern = r'var ytInitialPlayerResponse = ({.*?});'
match = re.search(pattern,driver.page_source)

if match:
    data = json.loads(match.group(1))
    # Exact view count
    exact_views = data['videoDetails']['viewCount']
    # Exact upload date (ISO format)
    upload_date = data['microformat']['playerMicroformatRenderer']['uploadDate']
    print("Exact Views:", exact_views)
    print("Upload Date:", upload_date)
else:
    print("Could not find the JSON data.")

driver.quit()
```

### expression systems used in programming for pattern matching, logic, and data extraction.

#### **Here Few Important Ones For Web Scrapping Purpose**

**Regex:**

When you have raw text and need to extract numbers, dates, or custom patterns.

Example: "Views: 19,475,682" → extract 19475682.

**CSS Selector:**

When HTML has nice tags and classes.

Example: div.title, a[href].

**XPath:***

When HTML is messy or nested deeply.

Example: "Find the <span> inside the second <div> after a <p> that contains text 'Views'".



```python
text = soup.find_all('span' , class_="yt-core-attributed-string--link-inherit-color")[6].text
def clear_discription(text):
    #remove all the linr break
    text = text.replace('\n','')
    #removing all the unwanted parts from the discription
    junk = ['OUR CHANNELS', 'Follow us', 'Subscribe', 'http', 'More videos']
    for word in junk:
        ttx = text.find(word)
        if ttx != -1 :
            text = text[:ttx]
            break

    #Remove all the extra spaces
    text = ' '.join(text.split())

    return text
    
```
